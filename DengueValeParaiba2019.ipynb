# Importando Bibliotecas que seram utilizadas

```python
# Importando Lib Pandas
import pandas as pd

# Informar ao Matplotlib sobre os gráficos inline
%matplotlib inline

#Importando Lib Matplotlib
import matplotlib.pyplot as plt

#Importando lib Numpy
import numpy as np
from scipy import stats

#Importando lib Seaborn
import seaborn as sns

#Setando estilo no Seaborn
sns.set(style="dark")
```

# Carregando Dataset

```python
# abrindo o dataset com a lib Pandas
f= pd.read_csv('Planilha_Dengue2019.csv',sep=',', encoding='cp1252')
```

# Tratamento dos dados do Dataset

```python
#Tratando dados presentes no dataset, uma vez que todos vieram dtype 'object'

f['populacao'] = f['populacao'].astype('int64')

f['janeiro_notificados'] = f['janeiro_notificados'].astype('int64')
f['janeiro_autoctones'] = f['janeiro_autoctones'].astype('int64')
f['janeiro_importados'] = f['janeiro_importados'].astype('int64')

f['fevereiro_notificados'] = f['fevereiro_notificados'].astype('int64')
f['fevereiro_autoctones'] = f['fevereiro_autoctones'].astype('int64')
f['fevereiro_importados'] = f['fevereiro_importados'].astype('int64')

f['marco_notificados'] = f['marco_notificados'].astype('int64')
f['marco_autoctones'] = f['marco_autoctones'].astype('int64')
f['marco_importados'] = f['marco_importados'].astype('int64')

f['abril_notificados'] = f['abril_notificados'].astype('int64')
f['abril_autoctones'] = f['abril_autoctones'].astype('int64')
f['abril_importados'] = f['abril_importados'].astype('int64')

f['maio_notificados'] = f['maio_notificados'].astype('int64')
f['maio_autoctones'] = f['maio_autoctones'].astype('int64')
f['maio_importados'] = f['maio_importados'].astype('int64')

f['junho_notificados'] = f['junho_notificados'].astype('int64')
f['junho_autoctones'] = f['junho_autoctones'].astype('int64')
f['junho_importados'] = f['junho_importados'].astype('int64')

f['julho_notificados'] = f['julho_notificados'].astype('int64')
f['julho_autoctones'] = f['julho_autoctones'].astype('int64')
f['julho_importados'] = f['julho_importados'].astype('int64')

f['agosto_notificados'] = f['agosto_notificados'].astype('int64')
f['agosto_autoctones'] = f['agosto_autoctones'].astype('int64')
f['agosto_importados'] = f['agosto_importados'].astype('int64')

f['setembro_notificados'] = f['setembro_notificados'].astype('int64')
f['setembro_autoctones'] = f['setembro_autoctones'].astype('int64')
f['setembro_importados'] = f['setembro_importados'].astype('int64')

f['outubro_notificados'] = f['outubro_notificados'].astype('int64')
f['outubro_autoctones'] = f['outubro_autoctones'].astype('int64')
f['outubro_importados'] = f['outubro_importados'].astype('int64')

f['novembro_notificados'] = f['novembro_notificados'].astype('int64')
f['novembro_autoctones'] = f['novembro_autoctones'].astype('int64')
f['novembro_importados'] = f['novembro_importados'].astype('int64')

f['dezembro_notificados'] = f['dezembro_notificados'].astype('int64')
f['dezembro_autoctones'] = f['dezembro_autoctones'].astype('int64')
f['dezembro_importados'] = f['dezembro_importados'].astype('int64')

f['total_notificados'] = f['total_notificados'].astype('int64')
f['total_autoctones'] = f['total_autoctones'].astype('int64')
f['total_importados'] = f['total_importados'].astype('int64')
```

# Gráfico de Pizza - Indice de Notificados por Região

```python
totalReg = list(f.regiao.drop_duplicates())

plt.subplots(figsize=(10, 10))
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0)  
plt.pie(f.regiao.value_counts(), labels=totalReg, explode=explode, colors=colors,
autopct='%1.0f%%', shadow=True, startangle=1000)
plt.title("Indice de Notificados por Região")
plt.show()
```
